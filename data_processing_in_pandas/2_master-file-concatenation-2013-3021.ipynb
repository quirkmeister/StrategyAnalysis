```python
# Dependencies

import pandas as pd
import numpy as np
import glob
import os
import timeit                                # To check performance
from datetime import datetime as dt
```

```python
# Set year to concatenate

year_to_process = 2021
```

```python
# Concatenate csv files

tic = timeit.default_timer()                            # Monitor performance

path = f'/Volumes/GoogleDrive/My Drive/Humanate/Tec Bootcamp/Tableau_Citibike/0. raw_data/{year_to_process}'
all_files = glob.glob(path + "/*.csv")

csv_file_list = []

for filename in all_files:
    try:
        df = pd.read_csv(filename, index_col=None, header=0)
        csv_file_list.append(df)
#         print (f'File {i} added')
    except:
        print(f'Concatenate error {filename}, could not be added')
        pass

citibike_rides_df = pd.concat(csv_file_list, axis=0, ignore_index=True)

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to extract and concatenate excel files: {round(toc - tic, 2)}')

citibike_rides_df
```

```python
# Clean up data and add an Age column

tic = timeit.default_timer()                            # Monitor performance

citibike_rides_df['started_at'] = pd.to_datetime(citibike_rides_df['started_at'])          # Convert to datetime values  
citibike_rides_df['ended_at'] = pd.to_datetime(citibike_rides_df['ended_at'])

citibike_rides_df = citibike_rides_df.rename(columns={'started_at' : 'Start Time & Date',# Polish up column names
                                                      'ended_at' : 'End Time & Date',
                                                      'start_station_id': 'Start Station ID',
                                                      'start_station_name': 'Start Station Name',
                                                      'start_lat': 'Start Lat',
                                                      'start_lng': 'Start Lon',
                                                      'end_station_id': 'End Station ID',
                                                      'end_station_name': 'End Station Name',
                                                      'end_lat': 'End Lat',
                                                      'end_lng': 'End Lon',
                                                      'bikeid': 'Bike ID',
                                                      'member_casual': 'User Type'})

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to process data: {round(toc - tic, 2)}')

for column in citibike_rides_df.columns:
    print(f'{column}, type: {type(citibike_rides_df.loc[0, column])}')

citibike_rides_df.head()
```

```python
citibike_rides_df = citibike_rides_df.rename(columns={'start_station_name': 'Start Station Name',
                                                      'start_lng': 'Start Lon'})

citibike_rides_df['User Type'].replace(to_replace='casual', value='Customer', inplace=True )           # Replace user type values
citibike_rides_df['User Type'].replace(to_replace='member', value='Subscriber', inplace=True )
citibike_rides_df
```

```python
citibike_rides_df.drop(['ride_id', 'rideable_type' ], inplace=True, axis=1)
citibike_rides_df
```

```python
citibike_rides_df_test = citibike_rides_df.copy()
citibike_rides_df_test
```

```python
citibike_rides_df_test['Duration (Sec)'] = citibike_rides_df_test['End Time & Date'] - citibike_rides_df_test['Start Time & Date'] / np.timedelta64(1, 's')
citibike_rides_df_test
```

```python
citibike_rides_df_test['Gender'] = np.nan
citibike_rides_df_test['Bike ID'] = np.nan
citibike_rides_df_test['Birth Year'] = np.nan
citibike_rides_df_test
```

```python
citibike_rides_df_test['Age'] = np.nan
citibike_rides_df_test
```

```python
citibike_rides_df_test = citibike_rides_df_test[['Duration (Sec)', 
                                                 'Start Time & Date',
                                                 'End Time & Date',
                                                 'Start Station ID',
                                                 'Start Station Name',
                                                 'Start Lat',
                                                 'Start Lon',
                                                 'End Station ID',
                                                 'End Station Name',
                                                 'End Lat',
                                                 'End Lon',
                                                 'Bike ID',
                                                 'User Type',
                                                 'Birth Year',
                                                 'Gender',
                                                 'Age']]
citibike_rides_df_test
```

```python
citibike_rides_df2 = pd.read_csv("../0. raw_data/202101-citibike-tripdata.csv", index_col=None, header=0)
citibike_rides_df2
```

```python
# Clean up data and add an Age column

tic = timeit.default_timer()                            # Monitor performance

citibike_rides_df2['starttime'] = pd.to_datetime(citibike_rides_df2['starttime'])          # Convert to datetime values  
citibike_rides_df2['stoptime'] = pd.to_datetime(citibike_rides_df2['stoptime'])

citibike_rides_df2['birth year'] = pd.to_numeric(citibike_rides_df2['birth year'], errors='coerce')  # Convert to float

citibike_rides_df2['Age'] = pd.DatetimeIndex(citibike_rides_df2['starttime']).year - citibike_rides_df2['birth year'] # Calculate age

citibike_rides_df2['gender'].replace(to_replace=1, value='Male', inplace=True )           # Replace gender values
citibike_rides_df2['gender'].replace(to_replace=2, value='Female', inplace=True )
citibike_rides_df2['gender'].replace(to_replace=0, value='Unknown', inplace=True )

citibike_rides_df2 = citibike_rides_df2.rename(columns={'tripduration': 'Duration (Sec)',  # Polish up column names
                                                      'starttime' : 'Start Time & Date',
                                                      'stoptime' : 'End Time & Date',
                                                      'start_station_id': 'Start Station ID',
                                                      'start station name': 'Start Station Name',
                                                      'start station latitude': 'Start Lat',
                                                      'start station longitude': 'Start Lon',
                                                      'end station id': 'End Station ID',
                                                      'end station name': 'End Station Name',
                                                      'end station latitude': 'End Lat',
                                                      'end station longitude': 'End Lon',
                                                      'bikeid': 'Bike ID',
                                                      'usertype': 'User Type', 	
                                                      'birth year': 'Birth Year',
                                                      'gender':'Gender'})

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to process data: {round(toc - tic, 2)}')

for column in citibike_rides_df2.columns:
    print(f'{column}, type: {type(citibike_rides_df2.loc[0, column])}')

citibike_rides_df2.head()
```
